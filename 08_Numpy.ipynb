## EXERCICE

```python
def initialisation(m, n):
    A = np.random.randn(m, n)
    A = np.concatenate((A, np.ones((m,1))), axis = 1)
    return A

x = initialisation(3, 2)
print(x)
```

```python
x.shape
```

**CORRECTION**

```python
def initialisation(m, n):
    A = np.random.randn(m, n)
    A = np.concatenate((A, np.ones((A.shape[0],1))), axis = 1)
    return A

x = initialisation(3, 2)
print(x)
print(x.shape)
```

## EXERCICE

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
r = face.shape[0]
c = face.shape[1]
zoom_face = face[r//4:-r//4, c//4:-c//4]

plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
zoom_face[zoom_face>200] = 255
```

```python
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

**CORRECTION**

```python
zoom_face[zoom_face>150] = 255
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```
