```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.cos(x)
```

**SUBPLOTS**

```python
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y, '-r', label = '$cos(x)$')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x, x**3, '+g', lw=2.5)
plt.plot(x, x**3, '-k', label ='${x}^{3}$')
plt.legend()

plt.show()
```

```python
import numpy as np

x = np.linspace(-5, 5, 100)

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x,np.sinh(x),'2b', lw = 4, label = 'sinh(x)')
plt.plot(x,np.sinh(x),'-k', lw = 2)
plt.legend()

plt.subplot(222)
plt.plot(x,x**3,'-k', lw = 4, label ='${x}^{3}$' )
plt.legend()

plt.subplot(223)
plt.plot(x,np.log(x**2),'-m', lw = 2, label ='$ln({x}^{2})$' )
plt.legend()

plt.subplot(224)
plt.plot(x,-2*x**2+3*x-5,'vr', lw = 0.5, label ='$-2{x}^{2}+3x-5$' )
plt.plot(x,-2*x**2+3*x-5,'-k', lw = 2)
plt.legend()

plt.savefig('figure1.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

def graphique(data):
    plt.figure(figsize=(12, 8))
    n=len(data)
    for k, i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1, i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```
