## INTERPOLATION

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = x**2
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d

f = interp1d(x,y, kind='linear')

new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x,y)
plt.scatter(new_x, result, c='r')
```

## OPTIMISATION

```python
x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 3/5*x**2 + 2 + np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

```python
from scipy import optimize
```

```python
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

1. curve_fit

```python
params, params_cov = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x,y)
plt.plot(x, f(x,params[0], params[1], params[2], params[3]), c='g', lw=3)
```

2. minimize

```python
def f(x):
    return x**2 + 15*np.sin(x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
from scipy import optimize

result = optimize.minimize(f, x0=-5).x
x0=-5
```

```python
plt.plot(x, f(x), lw=3, zorder=-1)
plt.scatter(result, f(result), s=100, c='r', zorder=1)
plt.scatter(x0, f(x0), s=200, marker = '+', c='g', zorder=1)
plt.show()
```

```python
def f(x):
    return np.sin(x[0]) + np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
import numpy as np

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

X,Y = np.meshgrid(x,y)
Z = f(np.array([X,Y]))
```

```python
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z, cmap='plasma')
plt.show()
```

```python
plt.contour(X,Y, f(np.array([X,Y])), 20)
plt.show()
```

```python
from scipy import optimize

x0 = np.zeros((2,))
result = optimize.minimize(f, x0=x0).x

plt.contour(X,Y, f(np.array([X,Y])), 20)
plt.scatter(result[0], result[1], s=100, c='g', zorder=1)
plt.scatter(x0[0], x0[1], s=200, marker = '+', c='r', zorder=1)
plt.show()

print(f'le minimum est le point de coordonnes : {result}')
```
